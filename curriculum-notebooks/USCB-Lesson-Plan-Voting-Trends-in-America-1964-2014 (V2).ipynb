## 1.1 How is the Data Used?<a id='s1.1'></a>

The census form is a way for the government to get a good idea as to **who** it is serving, which is: **YOU!** The government uses data it gets from this nation-wide Census to determine funding distributions across U.S. communities, as well as understand where community services are needed and how to implement them. Because of the Census, your neighborhood and city can improve education and transportation, promote public health, and use given money to make improvements. More services can be given to the elderly, new roads and schools can be built, and more job training centers can be maintained and established. *Everyone* benefits.

## 1.2 Census and Congresional Represenation <a id='s1.2'></a>

The amount of representation that your state has changes within Congress (in the U.S. House of Representatives) every ten years, when the Census is distributed. The voices of those vouching for your state's well-being in Congress could change or stay the same. Additionally, the representative (or "voice") representing your community in local, state, and federal governments could change, because district lines are redrawn every time. Members of Congress, state legislators, and many county and municipal offices are elected by voters grouped into districts, and you could be voting for different people depending on how these lines are drawn. These representatives are your spokespeople - their job is to listen to your needs and support government policies that will help your community! 

For more information, check out this document [here](https://www2.census.gov/programs-surveys/sis/resources/census101.pdf), designed to inform students like you on why the Census is so important! And for more reading on the confusing topic of re-districting, read [here](https://www.brennancenter.org/analysis/7-things-know-about-redistricting).

# 2. Vocabulary

Here are terms you should be familiar with before getting started with the lesson plan. 

 • **Election** – a formal and organized process for choosing leaders and representatives for a group <br>
 • **Vote** – a formal way to choose between two or more people or things <br>
 • **Citizen** – a person who can vote in a specific state or nation <br>
 • **Suffrage** – the right to vote <br>
 • **Registered voter** – a person who signed up ahead of time to vote by proving citizenship <br>
 • **Presidential election** – an election, held every four years in the United States, to choose a president and
vice president <br>
 • **Congressional election** – an election, held every two years in the United States, to choose certain
members of Congress

The original Consitution left the issue of voting rights to individual states, but 6 ammendment helped inform state law. Here are a list of the 6 ammendments:

 • The **14th Amendment** said that men older than 21 can vote. <br>
 • The **15th Amendment** said that people of any race can vote. <br>
 • The **19th Amendment** gave women the right to vote. <br>
 • The **23rd Amendment** gave citizens in the District of Columbia the right to vote in presidential elections. <br>
 • The **24th Amendment** got rid of poll taxes in elections for federal officials. <br>
 • The **26th Amendment**  lowered the voting age to 18.

### Item 3:
Now let's look at **Item 3: Voting and Registration Graphs, Congressional Elections, 2014**. These graphs represent the percent of voting registered voters by age and sex. 

```python
#Run this code to see Item 3 by pressing 'Shift' + 'Enter'


def rotated(array_2d):
    list_of_tuples = zip(*array_2d) #add [::-1] to reverse order
    return [list(elem) for elem in list_of_tuples]


#AGE PLOTS
ageRanges = ["18-29", "30-39", "40-49", "50-59", "60-69", "70+"]
legend = ["Voted", "Registered, no vote", "Not Registered", "No Response"]

eighteen = [20, 27, 31, 22]
thirty = [33, 30, 20, 17]
fourty = [40, 27, 18, 15]
fifty = [48, 20, 15, 17]
sixty = [59, 17, 12, 12]
seventy = [59, 17, 12, 12]

combined = list(rotated([eighteen, thirty, fourty, fifty, sixty, seventy]))
xrange = np.arange(1, 7)

ages = plt.figure(1, figsize=(10, 7))
# Voted
plt.bar(x=xrange, height=combined[0], color='navy', label='Voted')
# Registered, no vote
plt.bar(x=xrange, height=combined[1], bottom=combined[0], color='magenta', label='Registered, no vote')
# Not Registered
plt.bar(x=xrange, height=combined[2], bottom=[i+j for i,j in zip(combined[0], combined[1])], color='turquoise', label='Not Registered')
# No Response
plt.bar(x=xrange, height=combined[3], bottom=[i+j+k for i,j,k in zip(combined[0], combined[1], combined[2])], color='red', label='No Response')
plt.legend()
plt.title('Voting and registration by age in United States: 2014')
plt.xticks(xrange, ageRanges)
plt.ylabel('%')
ages.show()


#GENDER PLOTS
male = [40, 23, 20, 17]
female = [42, 25, 16, 17]

combined2 = list(rotated([male, female]))
xrange2 = np.arange(1, 3)

genders = plt.figure(2, figsize=(10, 7))
# Voted
plt.bar(x=xrange2, height=combined2[0], color='navy', label='Voted', figure=genders)
# Registered, no vote
plt.bar(x=xrange2, height=combined2[1], bottom=combined2[0], color='magenta', label='Registered, no vote', figure=genders)
# Not Registered
plt.bar(x=xrange2, height=combined2[2], bottom=[i+j for i,j in zip(combined2[0], combined2[1])], color='turquoise', label='Not Registered', figure=genders)
# No Response
plt.bar(x=xrange2, height=combined2[3], bottom=[i+j+k for i,j,k in zip(combined2[0], combined2[1], combined2[2])], color='red', label='No Response', figure=genders)
plt.legend()
plt.title('Voting and registration by sex in United States: 2014')
plt.xticks(xrange2, ['Male', 'Female'])
plt.ylabel('%')
genders.show()
```
